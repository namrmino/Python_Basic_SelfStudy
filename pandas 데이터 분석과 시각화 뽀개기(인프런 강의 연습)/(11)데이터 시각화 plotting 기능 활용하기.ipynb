# 19강 파이썬 데이터 시각화 히스토그램과 도수분포표 이해하기, 막대그래프 vs 히스토그램 차이점 이해하기

## Histograms(히스토그램)

### 도수분포표 vs 히스토그램

* 도수분포표 : 특정 구간에 속하는 자료의 개수를 표현
* 히스토그램 : 도수분포표를 시각화하여 막대그래프로 표현, 하지만 bar plot과는 다름

### 막대그래프 vs 히스토그램 (bar plot vs hist plot)

* bar plot : 표현 값에 비례하여 높이와 길이를 지닌 직사각형 막대로 범주형 데이터를 표현하는 차트나 그래프, 합계, 평균 등의 수치를 시각화
* hist plot : 구간별 빈도수를 표현

### 히스토그램 vs 정규분포 (hist plot vs density plot)

* hist plot : 구간별 빈도수
* density plot : 확률 밀도 함수 - 확률 변수의 밀도를 표현


* https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html#histograms

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline

# 한글폰트 사용시 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus']=False
```

```python
df4 = pd.DataFrame({'a' : np.random.randn(1000) + 1, 
                    'b' : np.random.randn(1000), 
                    'c' : np.random.randn(1000) - 1},
                  columns=['a','b','c'])
df4
```

```python
# alpha : 투명도
df4.plot.hist(alpha=0.5)
```

```python
# bins : 정사각형 픽셀 갯수
df4.plot.hist(stacked=True, bins=20)
```

```python
df4['a'].plot.hist(orientation='horizontal', cumulative=True)
```

# 20강 파이썬 판다스로 차분 diff 값을 구하고 히스토그램으로 표현하기

```python
# diff : 차분 값 => 앞의 값에서 뒤의 값을 뺀 값
df4['a_diff'] = df4['a'].diff()

df4['a_shift'] = df4['a'].shift()
df4['a_minus'] = df4['a'] - df4['a_shift']
df4[['a','a_shift', 'a_minus', 'a_diff']]
```

```python
# 차분값 히스토그램으로 그리기
df4['a'].diff().hist()
```

```python
df4[['a','b','c']].diff().hist(color='k', alpha=0.5, bins=50)
```

```python
data = pd.Series(np.random.randn(1000))
data
```
