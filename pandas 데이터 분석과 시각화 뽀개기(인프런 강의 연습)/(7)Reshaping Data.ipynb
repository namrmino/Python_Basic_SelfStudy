* ### melt와 pivot으로 Tidy Data 만들기
### 데이터 전처리할 때 많이 사용한다.

```python
df = pd.DataFrame({'A': {0: 'a', 1: 'b', 2: 'c'},
                   'B': {0: 1, 1: 3, 2: 5},
                   'C': {0: 2, 1: 4, 2: 6}})
df
```

```python
# 열 데이터를 행 데이터로 바꾸기
pd.melt(df, id_vars=['A'], value_vars=['B','C'])
```

```python
pd.melt(df, value_vars=['A','B','C']).rename(columns={
    'variable' : 'var',
    'value' : 'val'})
```

```python
df2 = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two', 'two'],
                   'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'baz': [1, 2, 3, 4, 5, 6],
                   'zoo': ['x', 'y', 'z', 'q', 'w', 't']})
df2
```

```python
# 행 데이터를 열 데이터로 바꾸기
df3=df2.pivot(index='foo', columns='bar', values='baz').reset_index()
```

```python
df3
```

```python
# pivot으로 바뀐 데이터를 melt로 다시 되돌리기
df3.melt(
    id_vars=['foo'],
    value_vars=['A','B','C']).sort_values(
    ['foo','bar']).rename(columns={'value':'baz'})
```

* ### 서로 다른 데이터를 하나의 데이터로 합쳐주는 &
 ### 여러 데이터를 하나의 데이터로 합쳐주는
 ### concat 사용하기

```python
s1 = pd.Series(['a', 'b'])
s1
```

```python
s2 = pd.Series(['c', 'd'])
s2
```

```python
pd.concat([s1, s2])
```

```python
pd.concat([s1, s2], ignore_index=True)
```

```python
pd.concat([s1, s2], keys=['s1','s2'])
```

```python
pd.concat([s1, s2], keys=['s1','s2'],names=['Series name','Row ID'])
```

```python
df1 = pd.DataFrame([['a', 1], ['b', 2]], columns=['letter', 'number'])
df1
```

```python
df2 = pd.DataFrame([['c', 3], ['d', 4]], columns=['letter', 'number'])
df2
```

```python
pd.concat([df1, df2])
```

```python
df3 = pd.DataFrame([['c', 3, 'cat'], ['d', 4, 'dog']],
                   columns=['letter', 'number', 'animal'])
df3
```

```python
pd.concat([df1, df3], sort=False)
```

```python
# 컬럼명이 다르거나 데이터가 다른 데이터를 합칠 때 join 옵션 사용
pd.concat([df1, df3], join="inner")
```

```python
df4 = pd.DataFrame([['bird', 'polly'], ['monkey', 'george']],
                   columns=['animal', 'name'])
df4
```

```python
pd.concat([df1, df4], axis=1)
```

```python
df5 = pd.DataFrame([1], index=['a'])
df5
```

```python
df6 = pd.DataFrame([2], index=['a'])
df6
```

```python
# 합칠 때 중복된 값이 있는지 확인할 때 verify_integrity 옵션을 사용한다.
pd.concat([df5, df6], verify_integrity=True)
```
